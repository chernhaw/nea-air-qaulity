# Find correlation between windspeed and PSI

```python
import pandas as pd
import json
import os

current_dir = os.getcwd()
print("Current directory:", current_dir)
```

## Load wind speed for 

```python
df_windspeed_2023 = pd.read_csv("HistoricalWindSpeedacrossSingapore2023.csv")
df_windspeed_2023.head()
```

## load psi for 2023

```python
df_psi_2023= pd.read_csv("HistoricalPollutantStandardsIndexPSI2023.csv")

df_psi_2023.head()
```

```python
df_psi_2023_central = df_psi_2023[df_psi_2023.region=="central"]
df_psi_2023_central.head()
```

## Get reading from central Singapore for PSI Ang Mo Kio 

```python
windspeed_amk= df_windspeed_2023[df_windspeed_2023.station_name=="Ang Mo Kio Avenue 5"]
windspeed_amk.head()
```

```python
windspeed_amk = windspeed_amk[['date', 'timestamp', 'reading_value']]
windspeed_amk.head()
```

```python
# Merge on 'timestamp' column, which is present in both dataframes
merged_df = pd.merge(df_psi_2023_central, windspeed_amk, on='timestamp', how='inner', suffixes=('_psi', '_wind'))
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=[
    'update_timestamp', 'region', 'region_id', 'location_longitude', 'location_latitude', 'date_wind'
])
merged_df.head()
```

## Find correlationship between wind speed and pollutant

```python
correlation = merged_df[['so2_sub_index', 'reading_value']].corr(method='pearson')
print(correlation)
```

```python
correlation = merged_df[['pm10_sub_index', 'reading_value']].corr(method='pearson')
print(correlation)
```

```python
correlation = merged_df[['o3_sub_index', 'reading_value']].corr(method='pearson')
print(correlation)
```

```python
correlation = merged_df[['pm25_sub_index', 'reading_value']].corr(method='pearson')
print(correlation)
```
